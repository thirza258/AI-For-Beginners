```python
feature_columns = [col for col in df_dosage.columns if col != target_column]
```

```python
X = df_dosage[feature_columns]
y = df_dosage[target_column]
```

```python
print(f"\nFeatures (X) head:\n{X.head()}")
print(f"\nTarget (y) head:\n{y.head()}")
```

```python
numerical_features = X.select_dtypes(include=np.number).columns.tolist()
categorical_features = X.select_dtypes(include='object').columns.tolist()
```

```python
if X[categorical_features].isnull().any().any():
    print(f"\nNaNs found in categorical features: {categorical_features}. Imputing with mode.")
    for col in categorical_features:
        if X[col].isnull().any():
            X[col] = X[col].fillna(X[col].mode()[0])
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_features),
        ('cat', OneHotEncoder(handle_unknown='ignore', sparse_output=False), categorical_features)
    ],
    remainder='passthrough'
)
```

```python
model = RandomForestClassifier(random_state=42)
```

```python
pipeline = Pipeline(steps=[('preprocessor', preprocessor),
                           ('classifier', model)])
```

```python
pipeline.fit(X, y)
```

Classification Model

```python
import os
import pandas as pd
from typing import Literal
from pydantic import BaseModel, Field
# Import LangChain components using the newer recommended structure
from langchain_core.output_parsers import PydanticOutputParser
from langchain_core.prompts import PromptTemplate
from langchain_core.runnables import Runnable
from langchain_openai import OpenAI

# --- Define the Pydantic Model and Parser ---
FormulationCategory = Literal[
    "Low Dosage",
    "Mid Dosage",
    "High Dosage"
]

class DosageResponse(BaseModel):
  dosage_category: FormulationCategory = Field(description="The dosage category this patient profile is classified into (Low, Mid, or High Dosage), based on the described supervised ML logic.")
  reasoning: str = Field(description="Explanation of why this patient profile is assigned to this category. Describe how a supervised model trained on patient clusters and effective formulations (using the provided data as examples) would arrive at this decision based on the input profile.")
  dosage: str = Field(description="The specific dosage value (e.g., '200 mg') derived from the provided data that seems most relevant for this patient profile and category. If no direct match, infer a suitable value based on patterns.")
  confidence_score: float = Field(ge=0.0, le=1.0, description="A simulated confidence score (0.0-1.0) for this classification, indicating how well the input profile matches patterns or examples in the provided dosage data.")

# --- Define the main classification function ---
# Added df_dosage as an argument
# Added type hints for clarity
def ai_dosage_recommendation_classification(input_patient_profile: dict | str, df_dosage: pd.DataFrame) -> DosageResponse:
    """
    Classifies a patient profile into a dosage category simulating a supervised ML model.

    Args:
        input_patient_profile: A dictionary or string describing the patient profile.
                               (A dictionary is recommended for structured input).
        df_dosage: A pandas DataFrame containing historical dosage data, including
                   features (patient/mushroom details) and targets (dosage, frequency, etc.).

    Returns:
        A DosageResponse object containing the classification result.
    """

    parser = PydanticOutputParser(pydantic_object=DosageResponse)

    # Convert DataFrame to a string format suitable for the prompt
    # Using to_string() is better than just putting the df object directly
    dosage_data= df_dosage
    target_column = ['Dosage (Optimal Concentration) [mg]','Frequency of Dosage',	'Duration of Use', 'Component'	,'Mushroom Species']
    feature_columns = [col for col in dosage_data.columns if col not in target_column]
    dosage_data_string = dosage_data.to_string()

    pdf_snippet_for_context = """
      Classification models (supervised ML) can be trained on labeled data (where labels are the
      formulation category that helped a given patient cluster) to automatically classify a new patient
      into one of the known formulation categories. With a growing dataset from Replenish’s tool
      (patients’ reported symptoms and outcomes), the classifier can continuously improve. Even with
      limited initial data, a simple rule-based or nearest-neighbor classification using the curated trial
      evidence could serve until more training data accumulates.
    """

    prompt_template_string = """
    You are simulating the behavior of a supervised Machine Learning classification model.

    Your purpose is to classify a given patient profile into a known "Formulation Category" (Low Dosage, Mid Dosage, or High Dosage).

    This classification should simulate the logic of a model trained on historical data (provided below) where patient clusters were associated with effective dosage formulations and categories. The model would learn the patterns between patient characteristics (features) and effective dosage categories and values (targets).

    **Context on the Simulated ML Model:**
    {pdf_context}

    **Historical Research Data (Simulated Training Data):**
    Based on this data, identify patterns linking characteristics (like mushroom species, component, frequency, duration, and any implied patient characteristics) to the effective dosage categories and specific dosage values.
    {research_data}

    **Patient Profile to Classify:**
    Analyze this patient profile:
    {patient_input}

    **Instructions:**
    1. Analyze the `patient_input` and compare it to the patterns and examples found in the `research_data`.
    2. Simulate the classification process: determine which "Formulation Category" (Low, Mid, or High Dosage) the `patient_input` would likely be assigned to by a supervised model trained on this data.
    3. Identify a relevant `dosage` value (e.g., '200 mg') from the `research_data` or infer a suitable value based on the patterns for the chosen category and the patient profile.
    4. Provide a `reasoning` that explains the classification decision, referencing how a model would use the provided data and the input profile.
    5. Assign a simulated `confidence_score` (0.0 to 1.0) based on how clear the match is between the input profile and the patterns in the research data.

    {format_instructions}
    """

    # --- Instantiate PromptTemplate ---
    # Use the template string and specify the input variables it expects
    prompt = PromptTemplate(
        template=prompt_template_string,
        input_variables=["pdf_context", "research_data", "patient_input"],
        partial_variables={"format_instructions": parser.get_format_instructions()} # Pass format instructions here
    )

    # --- Instantiate LLM ---
    # Make sure your OPENAI_API_KEY environment variable is set
    llm = OpenAI(temperature=0.1)

    # --- Create and Run the Chain ---
    # Use LCEL pipeline syntax: prompt | llm | parser
    chain: Runnable = prompt | llm | parser

    # Execute the chain by invoking it with the required input variables
    response: DosageResponse = chain.invoke({
        "pdf_context": pdf_snippet_for_context,
        "research_data": dosage_data_string,
        "patient_input": input_patient_profile
    })

    # Return the parsed response object
    return response
```

```python
patient_profile_example = {
        "Patient Description": "Adult patient seeking help with chronic inflammation.",
        "Age": 45,
        "Health Status": "Generally healthy, but chronic inflammation.",
        "Desired Outcome": "Reduce inflammation."
    }

try:
    print("Running classification for patient profile...")
    recommended_dosage = ai_dosage_recommendation_classification(
        input_patient_profile=patient_profile_example,
        df_dosage=df_dosage
    )

    print("\n--- Dosage Recommendation ---")
    print(f"Dosage Category: {recommended_dosage.dosage_category}")
    print(f"Recommended Dosage: {recommended_dosage.dosage}")
    print(f"Reasoning: {recommended_dosage.reasoning}")
    print(f"Simulated Confidence: {recommended_dosage.confidence_score:.2f}")

except Exception as e:
    print(f"\nAn error occurred: {e}")
    print("Please ensure:")
    print("- Your OPENAI_API_KEY environment variable is set correctly.")
    print("- The necessary libraries (langchain-openai, pydantic, pandas) are installed.")
    print("- You have billing set up for the OpenAI API.")
```
